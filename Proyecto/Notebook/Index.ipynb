# Introducción

Este documento contiene la memoria de los ejercicios de la asignatura Visión Artificial del cuarto curso del grado en Ingeniería Informática de la Universidad de Murcia realizados por Adrián Cánovas Rodríguez.

En este archivo se encuentran los enlaces y la guía de lectura de las practicas, así como una introducción, un análisis de los directorios y una conclusión.

Todos los arhivos mencionados se incluyen con este documento así como los archivos fuente de generación de estos notebooks.

## Archivos y carpetas

El directorio de prácticas se divide en diferentes carpetas donde pueden encontrarse los archivos relativos a cada ejercicio. Para cada uno de ellos existe un notebook asociado y generado en html (precompilado). Por tanto veamos la jerarquía de directorios de este proyecto. Cabe destacar que los apartados tienen el nombre y numeración de cada ejercicio propuesto en el enunciado de estas prácticas.

- Documentación: Esta carpeta contiene los notebook precompilados en formato HTML para su lectura
- Images: Contiene todos los recursos visuales (videos e imágenes) generados tanto para la realización de los ejercicios como para la de la memoria
- Notebook: Los notebook sin compilar realizados para cada ejercicio
- Source: Contiene todo el código generado para cada ejercicio

La unica salvedad en relación a la nomenclatura de carpetas y archivos es la carpeta de imágenes `chessboard` correspondiente al primer ejercicio. Esto es debido a que no considero la carpeta parte del ejercicio pues no se incluye el script para evaluarlo debido a que es el mismo que nos aporta el profesor en clase.

## Metodología de trabajo

Estas prácticas se han realizado siguiendo de forma guiada los boletines y notebooks relacionados con la asignatura. Sin embargo se ha consultado numerosas páginas que realizan las mismas acciones utilizando otras librerías y funciones y en ocasiones el código se ha visto afectado por esta implementación. El trabajo autonomo prima en estos ejercios.

Se ha utilizado un repositorio de git para la sincronización con un proyecto centralizado mientras se ha realizado su desarrollo en Mac (ordenador de casa) y Windows 10 (ordenador portatil que me llevaba al trabajo). Cabe destacar que algunas soluciones implementadas en Mac no han funcionado en windows aunque esto ha sido en contadas ocasiones casi siempre es por un tema de librerías o por paralelización.

En cuanto al entorno de desarrollo no se ha utilizado Anaconda sino PyCharm, esto se debe a que durante el primer cuatrimestre comencé a realizar el TFG desplegando algoritmos de machine learning en python utilizando este programa. Los entornos virtuales utilizados me permitian plena portabilidad (o casi) entre mis dos ordenadores lo que ha facilitado la realización de la práctica.

## Conclusión y opinión personal

En general es una asignatura muy interesante con muchas aplicaciones y que te permite investigar y "trastear" tanto como desees. Sin embargo, este año por temas de tiempo y circunstancias me ha parecido una asignatura relegada a un segundo plano: Cuando otras asignaturas se han llevado casi todo mi tiempo y dedicación para poder sacarlas VIA ha sido una asignatura que no presenta una dificultad muy alta y eso ha hecho en mi caso que el poco tiempo que tengo se lo haya tenido que dedicar a otras asignaturas para poder aprobarlas.

Me habría gustado disponer de entregas de evaluación anticipada que nos permita y obligue a realizar ejercicios con antelación para ir superando poco a poco la asignatura.

En mi caso personal el último ejercicio ha sido el que más se me ha atascado y por temas de tiempo no he podido pedir ayuda, he tenido que buscar información y otros ejercicios parecidos para poder realizarlo. El tema me interesa, pero como he dicho, no he tenido tiempo de ponerme lo suficiente.

A titulo personal dispongo de varios bots de telegram funcionando actualmente 24/7. En uno de ellos se gestiona una liga de futbol aficionado donde cada jugador tiene una ficha y puede comentar en el grupo para apuntarse a partidos, para crear un partido o para indicar los goles que ha marcado en una jornada.

Por otro lado, tengo una raspberry pi y me gustaría este verano poder lanzar algoritmos de machine learning (no muy exigentes) para realizar tratamiento y clasificación de imágenes. Por tanto, considero que esta asignatura es muy importante y seguiré aprendiendo sobre el tema más allá de esta.
